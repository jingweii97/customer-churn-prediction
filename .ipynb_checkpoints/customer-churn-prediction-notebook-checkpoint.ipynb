# Data Preprocessing for Baseline Model

For initial iteration of baseline model, data will be preprocessed by scaling and encoding the features accordingly. 

Feature engineering will be introduced at later iteration to see if it improves the baseline model performance if necessary.

## Feature Scaling and Encoding

```python
# Separate the features into numerical and categorical grouping
cat_col = list(strat_train_set.select_dtypes(include = 'O').columns)
num_col = list(strat_train_set.select_dtypes(include = np.number).columns)

# Remove target from cat_col
if 'Churn' in cat_col:
    cat_col.remove('Churn')

print(f'Categorical col: {cat_col}')
print(f'Numerical col: {num_col}')
```

For the numerical features
- 'SeniorCitizen' is actually categorical attributes which have already been encoded, hence no need for 
- 'MonthlyCharges' and 'TotalCharges' are positively skewed data, will be log transformed before scaling.
- 'tenure' is U-shaped distribution, hence will be binned according to Short,Medium and Long tenure.

```python
# Rearrange the feature group accordingly
log_col = ['MonthlyCharges', 'TotalCharges']
bin_col = ['tenure']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import FunctionTransformer, KBinsDiscretizer, StandardScaler, OneHotEncoder

# Define the preprocessing steps
log_scaler_transformer = Pipeline(steps = [('log',FunctionTransformer(np.log1p, feature_names_out='one-to-one')),
                                          ('scale',StandardScaler())])

bin_transformer = KBinsDiscretizer(n_bins = 3, encode = 'onehot-dense' ,strategy='uniform',subsample=None)

encoder = OneHotEncoder(sparse_output = False)

preprocessor = ColumnTransformer(transformers = [('log_scaler',log_scaler_transformer,log_col),
                                                ('bin',bin_transformer,bin_col),
                                                ('encode',encoder,cat_col)],remainder = 'passthrough', verbose_feature_names_out=False)
```

## Fit and Transform Data

```python
# Drop target variable
train_X = strat_train_set.drop('Churn', axis=1)
train_y = strat_train_set['Churn']

# Peek at the fit and transformed data
transformed_train =  preprocessor.fit_transform(train_X)
transformed_train
# transformed_train_df = pd.DataFrame(transformed_train,
#                                     columns=preprocessor.get_feature_names_out(),
#                                    index=train_X.index)
# transformed_train_df.head()
```

# Model Evaluation #1

As mentioned, goal of this notebook is to correctly predict as much customer churn as possible, hence target is to achieve a 80% or more recall score while maintaining precision above 50%. With the mentioned target score, aim is to focus on high recall while maintaining reasonable precision.

```python
# Evaluation function
recall_target = 0.8

from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.metrics import ConfusionMatrixDisplay, make_scorer, recall_score, precision_recall_curve
def show_cross_val_recall(model, X, y, predict_method):
    print(f'{model["model"]}:')
    # # Get recall score using cross_val_score
    # recall_scorer = make_scorer(recall_score,pos_label='Yes')
    # cross_val_recall_score = cross_val_score(model,X,y, cv = 10, scoring = recall_scorer)

    # Get Confusion matrix and its recall score
    y_cross_val_pred = cross_val_predict(model,X,y, cv = 10)
    ConfusionMatrixDisplay.from_predictions(y,y_cross_val_pred)
    plt.show()
    # recall_score_from_cm = recall_score(y,y_cross_val_pred,pos_label='Yes')
    # print(f'Recall score using cross_val_predict {np.round(recall_score_from_cm,4)}')

    # Get Precision and recall vs threshold graph
    y_scores = cross_val_predict(model,X,y, cv = 10, method = predict_method)
    if predict_method == 'predict_proba':
        y_scores = y_scores[:,1]
        
    precisions,recalls,thresholds = precision_recall_curve(y,y_scores,pos_label='Yes')
    # Get the threshold of target score
    desire_idx = np.where(recalls > recall_target)[0]
    best_idx = desire_idx[np.argmax(precisions[desire_idx])]
    best_threshold = thresholds[best_idx] if best_idx < len(thresholds) else thresholds[-1]
    best_recall = recalls[best_idx]
    best_precision = precisions[best_idx]
    
    plt.plot(thresholds,precisions[:-1],label = "Precision")
    plt.plot(thresholds,recalls[:-1],label = "Recall")
    plt.vlines(best_threshold,0,1.0,'k',linestyles = 'dotted',label = 'Threshold')
    plt.scatter(best_threshold,best_recall,color = 'red',marker = 'o',label = 'Best recall')
    plt.text(best_threshold,best_recall,f'({best_threshold:.2f},{best_recall:.2f})')
    plt.scatter(best_threshold,best_precision,color = 'green',marker = 'o',label = 'Best precision')
    plt.text(best_threshold,best_precision,f'({best_threshold:.2f},{best_precision:.2f})')
    plt.legend(loc = 'best')
    plt.xlabel("Threshold")
    plt.show()

    # Get precision vs recall
    plt.plot(recalls,precisions,label = 'Precision/Recall curve')
    plt.scatter(best_recall,best_precision,color = 'red',marker = 'o',label = 'Best point')
    plt.text(best_recall,best_precision,f'({best_recall:.2f},{best_precision:.2f})')
    plt.legend(loc = 'best')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Precision vs Recall Curve')
    plt.show()
    
    print(f'best_threshold {best_threshold:.2f}')
    print(f'best_recall {best_recall:.2f}')
    print(f'best_precision {best_precision:.2f}')
    # print(f'Recall score using cross_val_score {cross_val_recall_score}')
```

```python
from sklearn.linear_model import LogisticRegression

# Model 1
log_reg = Pipeline(steps = [('preprocessor',preprocessor), ('model',LogisticRegression(random_state=SEED))])
show_cross_val_recall(log_reg,train_X,train_y,'decision_function')
```

```python
from sklearn.tree import DecisionTreeClassifier

# Model 2
tree_classifier = Pipeline(steps = [('preprocessor',preprocessor), ('model',DecisionTreeClassifier(random_state=SEED))])
show_cross_val_recall(tree_classifier,train_X,train_y, 'predict_proba')
```

```python
from sklearn.ensemble import RandomForestClassifier

# Model 3
forest_classifier = Pipeline(steps = [('preprocessor',preprocessor), ('model',RandomForestClassifier(random_state=SEED))])
show_cross_val_recall(forest_classifier,train_X,train_y,'predict_proba')
```

Among all the models, logistic regression performed the best, which achieved 80% recall with 52% precision, already fulfilled the target performance set earlier. However, more efforts will be invested to see if model can be improved further.

# Optimation #1: Feature Engineering

Based on previous EDA findings, there are several feature engineering that could be done to improve the model predictive power on churn rate, for example feature creation (generating new feature to highlight important patterns), feature selection (removing irrelevant features) and etc

## Feature Creation

```python
# Create several feature highlighting the important patterns found in previous EDA

churn_data_copy = churn_data.copy()


high_monthly_cost_threshold = np.percentile(churn_data_copy['MonthlyCharges'], 75)
low_total_charges_threshold = np.percentile(churn_data_copy['TotalCharges'], 25)

churn_data_copy['HighCost_NoCommitment'] = ((churn_data_copy['MonthlyCharges'] < high_monthly_cost_threshold) & (churn_data_copy['Contract'] == 'Month-to-month')).astype(int)
churn_data_copy['NoOnlineServiceAndProtection'] = ((churn_data_copy[['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport']] == 'No').all(axis = 1)).astype(int)
churn_data_copy['LowTotalCharges'] = (churn_data_copy['TotalCharges'] < low_total_charges_threshold).astype(int)
churn_data_copy['TotalHouseholdCount'] = (churn_data_copy['Partner'] == 'Yes').astype(int) +  (churn_data_copy['Dependents'] == 'Yes').astype(int)


churn_data_num = churn_data_copy.select_dtypes(include = np.number)
corr_matrix = churn_data_num.corr()
print(f"Correlation:"+ '\n' +f"{corr_matrix['Churn_Num'].sort_values(ascending = False)}")

churn_data_copy.head()
```

Correlation matrix shows that the new features have high correlation with the churn number, hence will be added to pipeline.

```python
# In previous EDA results show that some features don't contribute to predicting churn rate, hence will be removed to reduce noise
# These features are proved that they don't influence churn
irrelevant_features = ['gender','PhoneService']
```

## Transformer Update

Custom transformer is created to cater for feature creation, whereas feature selection can be achieved by simply dropping the irrelevant features.

```python
from sklearn.base import BaseEstimator, TransformerMixin

# Create feature creation transformer
class FeatureCreationTransformer(BaseEstimator,TransformerMixin):
    def fit(self,X, y = None):

        return self
        
    def transform(self, X):
        X = X.copy()
        high_monthly_cost_threshold = np.percentile(X['MonthlyCharges'], 75)
        low_total_charges_threshold = np.percentile(X['TotalCharges'], 25)
        # Vectorized operations
        X['HighCost_NoCommitment'] = ((X['MonthlyCharges'] < high_monthly_cost_threshold) & (X['Contract'] == 'Month-to-month')).astype(int)
        X['NoOnlineServiceAndProtection'] = ((X[['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport']] == 'No').all(axis = 1)).astype(int)
        X['LowTotalCharges'] = (X['TotalCharges'] < low_total_charges_threshold).astype(int)
        X['TotalHouseholdCount'] = (X['Partner'] == 'Yes').astype(int) +  (X['Dependents'] == 'Yes').astype(int)
        return X

    def get_feature_names_out(self, input_features=None):
        return [
            'HighCost_NoCommitment', 'NoOnlineServiceAndProtection',
            'LowTotalCharges', 'TotalHouseholdCount'
        ]
```

```python
# Remove the irrelevant col from cat col

# Create a new pipeline
cat_col_exc_irr_feat = [col for col in cat_col if col not in irrelevant_features]

preprocessor = ColumnTransformer(transformers = [('drop_col','drop',irrelevant_features),
                                                 ('log_scaler',log_scaler_transformer,log_col),
                                                ('bin',bin_transformer,bin_col),
                                                ('encode',encoder,cat_col_exc_irr_feat)],remainder = 'passthrough', verbose_feature_names_out=False)

new_pipeline = Pipeline([('feature_creation',FeatureCreationTransformer()),('preprocessor',preprocessor)])

transformed_train_X = new_pipeline.fit_transform(train_X)

transformed_train_df = pd.DataFrame(transformed_train_X,
                                    columns=new_pipeline['preprocessor'].get_feature_names_out(),
                                   index=train_X.index)
transformed_train_df.head()
```
